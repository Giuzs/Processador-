**1) Descreva as funções básicas de uma UCP, indicando os seus componentes principais.**

A UCP é responsável por controlar a operação do computador e realizar o processamento de dados. Suas funções principais são buscar as instruções na memória, decodificar o que deve ser feito e executar a operação correspondente.

Componentes principais:

Unidade de Controle (UC): Gerencia o funcionamento do computador, organizando a ordem em que tudo deve ser feito.

Unidade Aritmética e Lógica (UAL): Executa operações matemáticas e lógicas.

Registradores: Memórias internas para armazenamento temporário de dados e instruções.

**2) Quais são as funções da unidade aritmética e lógica – UAL?**

Executar operações aritméticas básicas (como adição, subtração, multiplicação e divisão).

Executar operações lógicas booleanas (como AND, OR, NOT, XOR).

Realizar deslocamentos de bits (à esquerda ou à direita) e comparações de valores.



**3) O que é e para que servia o ACC?**

O ACC significa Acumulador (Accumulator). É um registrador que guarda temporariamente os resultados das contas feitas pela UAL e também serve como um dos operandos de entrada para a próxima operação, minimizando a necessidade de acessos lentos à memória principal.

**5) Quais são as funções da unidade de controle de um processador?**

Buscar as instruções na memória principal.

Interpretar/Decodificar as instruções para determinar qual operação deve ser realizada.

Mandar comandos para coordenar o processador e as demais peças durante a execução da instrução.

Controlar o fluxo de dados internos e o relógio (clock) do sistema.

 **6) Seria possível realizar o projeto de um processador em que o tamanho em bits do CI fosse diferente do tamanho do REM? Nesse caso, qual dos dois registradores deveria ter maior tamanho? Por quê?**

Sim, é possível. O Rem deveria ser maior, pois o CI precisa apenas ser grande o suficiente para endereçar a região de memória onde o código do programa fica armazenado. Já o REM precisa ser capaz de endereçar o espaço total da memória principal, que inclui não apenas o código, mas também toda a área de dados e variáveis, exigindo, então, mais bits de endereçamento.

**7) Considere um computador cuja MP é organizada com N células de 1 byte cada uma. As instruções interpretadas pela UCP possuem três tamanhos diferentes: as do tipo A possuem 16 bits; as do tipo B têm 32 bits e as do tipo C possuem 48 bits. Considerando que o código de operação de cada uma tem um tamanho fixo e igual a 8 bits e que os programas executados nesse processor são constituídos de uma mistura dos três tipos de instruções, imagine um processo prático para incremento automático do CI após a execução de cada instrução de um programa.**

Como a memória é endereçada a byte, o incremento do CI deve depender do tamanho real da instrução que acabou de ser buscada.

No início do ciclo de busca, a UC lê o primeiro byte (8 bits) da instrução, que corresponde ao Código de Operação (OpCode). A partir da decodificação desse OpCode, a Unidade de Controle identifica se a instrução é do tipo A, B ou C.

Com base nessa identificação, o hardware incrementa o CI de forma variável:

Se for tipo A (16 bits = 2 bytes): CI <- CI + 2

Se for tipo B (32 bits = 4 bytes): CI <- CI + 4

Se for tipo C (48 bits = 6 bytes): CI <- CI + 6

**8) Considere um processador cujo ciclo de instrução não possua a etapa de incremento automático do valor do CI. Imagine um método alternativo que permita a execução do programa.**

Cada instrução do programa deveria conter, além dos seus operandos normais, um campo adicional indicando o endereço de memória da próxima instrução a ser executada.

Ao final da execução da instrução atual, a Unidade de Controle leria esse campo e carregaria o valor diretamente no CI.

**9) Qual é e onde se localiza o registrador cujo conteúdo controla a seqüência de processamento das instruções de um programa?**

O registrador é o CI (Contador de Instrução, também conhecido como PC - Program Counter). Ele localiza-se internamente na Unidade de Controle (UC) da UCP.

 **11) Qual é o registrador cujo conteúdo determina a capacidade de memória de um computador? Justifique.**

O REM (Registrador de Endereços de Memória), pois é ele que se conecta diretamente ao barramento de endereços. A quantidade de bits que o REM possui define o limite máximo de combinações binárias diferentes que o processador consegue gerar para referenciar uma célula de memória. Se o REM possui n bits, o processador só consegue endereçar no máximo $2^n$ locais distintos de memória.

**15) Quando se fala que um determinado microcomputador A é um micro de 8 bits e que um outro micro B é de 16 bits, a que estamos nos referindo? Ao tamanho da célula de MP ou ao tamanho da palavra? Qual a base desses dois conceitos (palavra e célula)?**

Estamos nos referindo ao tamanho da palavra do processador, pois um processador de 16 bits manipula palavras de 16 bits internamente, mas, ao acessar a Memória Principal, ele pode precisar ler duas células consecutivas de 8 bits (2 bytes) para formar uma única palavra.

**18) Explique a diferença entre um processamento seqüencial e um outro pipeline.**

A principal diferença está na forma como o tempo e o hardware são aproveitados para executar as instruções: o modelo sequencial faz uma coisa de cada vez do início ao fim, enquanto o pipeline sobrepõe etapas de instruções diferentes para ganhar velocidade.
